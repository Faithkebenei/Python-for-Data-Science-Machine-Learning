### The .merge() method
- Inner
- Outer
- Left or Right

### Inner

```python
registrations = pd.DataFrame({'reg_id':[1,2,3,4],'name':['Andrew','Bobo','Claire','David']})
logins = pd.DataFrame({'log_id':[1,2,3,4],'name':['Xavier','Andrew','Yolanda','Bobo']})
```

```python
registrations
```

```python
logins
```

```python
pd.merge(registrations,logins,how="inner",on="name")
```

### Left or Right

```python
pd.merge(registrations,logins,on="name",how="left")
```

```python
pd.merge(registrations,logins, on="name",how="right")
```

### Outer

```python
pd.merge(registrations,logins,on="name",how="outer")
```

```python
registrations = registrations.set_index("name")
```

```python
registrations
```

```python
logins
```

```python
pd.merge(registrations,logins,left_index=True,right_on="name",how="inner")
```

```python
registrations = registrations.reset_index()
```

```python
registrations.columns = ["reg_name","reg_id"]
```

```python
registrations
```

```python
logins
```

```python
results = pd.merge(registrations,logins,how="inner",left_on="reg_name",right_on="name")
```

### KDE plot
- KDE stands for Kernel Density Estimation.
- It is a method of estimating a probability density function of a random variable.

## Categorical plots
The categorical plots discussed here will display a statistical metrics per a category.
<br><br>
For example mean value per category or a count of the number of rows per category.
<br><br>
It is the visualization equivalent of a groupby() call.
<br><br>
The two main types of plots for this are:
- countplot() -Counts number of rows per category.
- barplot() -General form of displaying any chosen metric per category.

### Boxplot
- Displays the distribution of continuous variable.
- It does this through the use of quartiles.
- Quartiles separate out the data into 4 equal number of data points:
  - 25% of data points are in the bottom quartile.
